```python
import numpy as np
df_one_hot_encoded = pd.get_dummies(data = data1, columns =data1_objects)

print(df_one_hot_encoded)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

cols_to_encode = data1_objects
new_encoded_cols_names = []

for col in cols_to_encode:
    new_encoded_cols_names += [f"is_{category}" for category in data[col].unique().tolist()]

print(new_encoded_cols_names)


one_hot_encoder = OneHotEncoder(sparse = False, handle_unknown = 'ignore')
encoded_cols = one_hot_encoder.fit_transform(data[cols_to_encode])


df_encoded = pd.DataFrame(encoded_cols, columns = new_encoded_cols_names)
df_one_hot_encoded = data.join(df_encoded)

print(df_one_hot_encoded)
```

### Decision Tree 

##### Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems, but mostly it is preferred for solving Classification problems.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
parameter={
    'criterion':['gini','entropy'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
from sklearn.model_selection import GridSearchCV
dtc=DecisionTreeClassifier() #model object creation
dtc.fit(X_train,y_train) # training set
```

### Random forest

##### Random Forest is a popular machine learning algorithm that belongs to the supervised learning technique. It can be used for both Classification and Regression problems in ML. It is based on the concept of ensemble learning, which is a process of combining multiple classifiers to solve a complex problem and to improve the performance of the model.

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score  

rclf = RandomForestClassifier(n_estimators=100, max_features="log2",random_state=0)
rclf.fit(X_train,y_train)
# Fitting the Random Forest algorithm to the training set
```

```python
y_pred = rclf.predict(X_test)
y_pred
```

##### The confusion matrix to determine the correct and incorrect predictions

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy_score(y_test, y_pred)
```

### AdaBoost

```python
ada = AdaBoostClassifier(n_estimators=100)
ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
accuracy_score(y_test, y_pred)
```

### xgBoost

```python
import xgboost as xgb
xg_model = xgb.XGBClassifier()
xg_model.fit(X_train, y_train)
y_pred = xg_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

### Support vector classifier

```python
from sklearn.svm import SVC # "Support vector classifier"
svc = SVC(kernel = 'rbf', C = 3)
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
# Perform k-fold cross-validation (e.g., 5-fold)
k = 5
scores = cross_val_score(svc, X_train, y_train, cv=k, scoring='f1_macro')

# Calculate mean F1 score and standard deviation
mean_f1 = np.mean(scores)
std_f1 = np.std(scores)

print(f'F1 Score (Macro) for {k}-Fold Cross-Validation:')
print(f'Mean F1 Score: {mean_f1:.2f}')
print(f'Standard Deviation: {std_f1:.2f}')
```
